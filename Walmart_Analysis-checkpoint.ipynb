```python
#Changing the datatype of date
df['date'] = pd.to_datetime(df['date'], format = '%d/%m/%y')
```

```python
df['day'] = df['date'].dt.day
```

```python
df
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df
```

```python
df['quarter'] = df['date'].dt.quarter
```

```python
df
```

- I have changed the dtype of date in datetime dtype, Standarized the date format and extract day, month and quarter

```python
#Changing dtype of 'unit_price' column
df['unit_price'] = df['unit_price'].str.replace('$' , '')

df['unit_price'] = df['unit_price'].astype(float)

print(df['unit_price'].dtype)
```

- I have changed the dtype of 'unit_price' column into float also removed $ symbol

```python
#creating a new column 'total'
df['total'] = df['unit_price'] * df['quantity']
df.head(5)
```
